````python
import yfinance as yf
import pandas as pd

# Download historical stock price data for AAPL
ticker = "AAPL"
start_date = "2020-01-01"
end_date = "2023-01-01"

data = yf.download(ticker, start=start_date, end=end_date)

# Keep in memory and summarize key statistics
print(f"Summary Statistics for {ticker} ({start_date} to {end_date}):")
print(data.describe())
````

````python
# Step 2: Feature Engineering - Suggested Features

# 1. Moving Averages: Simple (SMA) and Exponential (EMA) for various windows (e.g., 5, 20, 50 days).
# 2. Technical Indicators: RSI (Relative Strength Index), MACD, and Bollinger Bands.
# 3. Volume Metrics: Volume change percentage and daily volume spikes.
# 4. Volatility Measures: Standard deviation of returns or Average True Range (ATR).
# 5. Lagged Features: Previous day's closing prices or returns.
# 6. Time-based Features: Day of the week, month, or quarter to capture seasonality.
# 7. Sentiment Data: (External) Sentiment scores from financial news or social media.
````

````python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import pandas as pd

# 1. Feature Engineering (using existing 'data' from previous steps)
data_clf = data.copy()
data_clf['MA_5'] = data_clf['Close'].rolling(window=5).mean()
data_clf['MA_20'] = data_clf['Close'].rolling(window=20).mean()
data_clf['Vol_Change'] = data_clf['Volume'].pct_change()

# Create Binary Target: 1 if tomorrow's Close > today's Close, else 0
data_clf['Target'] = (data_clf['Close'].shift(-1) > data_clf['Close']).astype(int)

# Drop rows with NaN values resulting from rolling/shifting
data_clf.dropna(inplace=True)

# Define features and target
features = ['Open', 'High', 'Low', 'Close', 'Volume', 'MA_5', 'MA_20', 'Vol_Change']
X = data_clf[features]
y = data_clf['Target']

# 2. Train/Test Split (70/30 ratio, shuffle=False to respect time series nature)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=False)

# 3. Model Building - Random Forest Classifier
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# 4. Evaluation
y_pred = rf_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print(f"Accuracy: {accuracy:.2%}")
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# 5. Feature Importance
importances = pd.DataFrame({
    'Feature': features,
    'Importance': rf_model.feature_importances_
}).sort_values(by='Importance', ascending=False)

print("\nFeature Importance:")
print(importances)
````

````python
import yfinance as yf
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# 1. Download new data for evaluation
eval_data = yf.download("AAPL", start="2023-01-01", end="2024-01-01")

# 2. Feature Engineering (must match the training features)
eval_df = eval_data.copy()
eval_df['MA_5'] = eval_df['Close'].rolling(window=5).mean()
eval_df['MA_20'] = eval_df['Close'].rolling(window=20).mean()
eval_df['Vol_Change'] = eval_df['Volume'].pct_change()

# Binary target for evaluation period
eval_df['Target'] = (eval_df['Close'].shift(-1) > eval_df['Close']).astype(int)

# Drop NaNs created by indicators
eval_df.dropna(inplace=True)

# Prepare features and target
# features list was defined in the previous step: ['Open', 'High', 'Low', 'Close', 'Volume', 'MA_5', 'MA_20', 'Vol_Change']
X_eval = eval_df[features]
y_eval = eval_df['Target']

# 3. Predict using the trained rf_model
y_eval_pred = rf_model.predict(X_eval)

# 4. Report Performance Metrics
accuracy_eval = accuracy_score(y_eval, y_eval_pred)

print(f"Evaluation Period: 2023-01-01 to 2024-01-01")
print(f"Accuracy: {accuracy_eval:.2%}")
print("\nConfusion Matrix:")
print(confusion_matrix(y_eval, y_eval_pred))
print("\nClassification Report:")
print(classification_report(y_eval, y_eval_pred))
````

````python
```python
import matplotlib.pyplot as plt

# Calculate daily returns for the evaluation period
eval_df['Daily_Return'] = eval_df['Close'].pct_change()

# Define the strategy:
# If prediction is 1 (Up), we hold the stock and get the daily return.
# If prediction is 0 (Down), we stay in cash (0 return).
# We shift the predictions because the model predicts 'tomorrow' based on 'today'.
eval_df['Strategy_Return'] = eval_df['Daily_Return'] * pd.Series(y_eval_pred, index=X_eval.index).shift(1)

# Fill NaN values created by shifting/pct_change
eval_df['Strategy_Return'] = eval_df['Strategy_Return'].fillna(0)
eval_df['Daily_Return'] = eval_df['Daily_Return'].fillna(0)

# Calculate Cumulative Returns
eval_df['Cumulative_Market_Return'] = (1 + eval_df['Daily_Return']).cumprod()
eval_df['Cumulative_Strategy_Return'] = (1 + eval_df['Strategy_Return']).cumprod()

# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(eval_df.index, eval_df['Cumulative_Market_Return'], label='Buy and Hold AAPL')
plt.plot(eval_df.index, eval_df['Cumulative_Strategy_Return'], label='RF Strategy', linestyle='--')
plt.title('Backtest: Random Forest Strategy vs. Buy and Hold (2023-2024)')
plt.xlabel('Date')
plt.ylabel('Cumulative Return (Multiplier)')
plt.legend()
plt.grid(True)
plt.show()

# Summary Statistics
total_market_return = (eval_df['Cumulative_Market_Return'].iloc[-1] - 1) * 100
total_strategy_return = (eval_df['Cumulative_Strategy_Return'].iloc[-1] - 1) * 100
print(f'Total Market Return: {total_market_return:.2f}%')
print(f'Total Strategy Return: {total_strategy_return:.2f}%')
```
````

- Potential Trading Strategies
Given the model's performance, potential trading strategies could be formulated as follows:

#### Conservative Strategy:

- **Focus**: Only trade on days when the model predicts a price increase with high confidence.
- **Action**: Buy stocks when the model predicts an increase (class 1).
- **Risk Management**: Since the model has a high precision for the no-increase class (class 0), it might be more reliable to stay out of the market when the model predicts no increase.

#### Contrarian Strategy:

- **Focus**: Take advantage of the model's high recall for no price increase predictions (class 0).
- **Action**: Sell short or avoid buying stocks on days when the model predicts no increase.
- **Risk Management**: This strategy assumes that the model’s predictions of no price increase are more reliable. It requires a careful assessment of market conditions to avoid false signals.

#### Hybrid Strategy:

- **Combine**: Use both conservative and contrarian strategies based on confidence thresholds.
- **Action**: Set a higher threshold for buying (only buy when the predicted probability of price increase is significantly high) and a lower threshold for selling or staying out of the market.
- **Risk Management**: Adjust thresholds dynamically based on recent model performance and market volatility.
